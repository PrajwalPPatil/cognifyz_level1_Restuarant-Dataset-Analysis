# 2.Check Missing values

```python
missing_values=data.isnull().sum()
print('missing values in each column:')
print(missing_values)
```

```python
#Fill missing values in "Cuisine" column with mode
Mode_cuisines=data['Cuisines'].mode()[0]
data['Cuisines'].fillna(Mode_cuisines,inplace=True)
print(f'Missing values in "Cuisines" column after filling with mode: {data["Cuisines"].isnull().sum()}')
```

# Analyze the Distribution of Target Variable

```python
#Target Variable:Aggregate rating
#Check data types
print(data['Aggregate rating'].dtypes)
```

```python
#Convert 'Aggregate rating' to numeric
data['Aggregate rating']=pd.to_numeric(data['Aggregate rating'],errors='coerce')
```

```python
#Analyze distribiution
plt.figure(figsize=(10,6))
sns.histplot(data['Aggregate rating'],kde=True,color='skyblue')
plt.title('Distribution of Aggregate Rating')
plt.xlabel('Aggregate Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
#Identify class imbalances
rating_counts=data['Aggregate rating'].value_counts()
print("Value counts of 'Aggregate rating':")
print(rating_counts)
```

# 3.Identifying Top cuisines & cities

# 1.Top cuisines

```python
#Calculate the top 10 most common cuisines
top_cuisines=data['Cuisines'].value_counts().head(10)
print("Top 10 Most Common Cuisines:\n",top_cuisines)
```

```python
#Visualization of  top cuisines

#Get the top 5 most common cuisines
top_cuisines=data['Cuisines'].value_counts().head(5)

plt.figure(figsize=(8,6))
sns.countplot(data=data[data['Cuisines'].isin(top_cuisines.index)],y='Cuisines',order=top_cuisines.index,palette="Set1")
plt.title('Top 5 Most Common Cuisines')
plt.xlabel('Count')
plt.ylabel('Cuisines')
plt.xticks(rotation=45)
plt.show()
```

# 2.Cities with Highest Number of Restuarants

```python
#Calculate the top 10  cities with the highest number of restuarants
top_cities=data['City'].value_counts().head(10)
#Convert the series to a DataFrame for better display
top_cities_data=top_cities.reset_index()
top_cities_data.columns=['City','Number of Restuarants']
#Display the DataFrame in tabular format
print("Top 10 Cities with Highest Number of Restuarants\n")
print(top_cities_data.to_string(index=False))
```

```python
#Visualization of Cities with the Highest number of Restuarants

#Get the top 5 Cities with the Highest number of Restuarants
top_cities=data['City'].value_counts().head(5)

plt.figure(figsize=(10,6))
sns.countplot(data=data[data['City'].isin(top_cities.index)],y='City',order=top_cities.index,palette="tab10")
plt.title('Top 5 Cities with the Highest Number of Restuarants')
plt.xlabel('Number of Restuarants')
plt.ylabel('City')
plt.xticks(rotation=45)
plt.show()
```

# Task3:Geospatial Analysis

# 1.Visualize Locations of Restuarants

```python
#Task.3]Geospatial Analysis
#1.Visualize location of restaurants
#Create a scatter map with additional customization
import plotly.express as px
fig = px.scatter_mapbox(data,lat='Latitude',lon='Longitude',hover_name='Restaurant Name',hover_data=['City','Cuisines','Aggregate rating','Price range','Votes'],
                        color='Aggregate rating',color_continuous_scale=px.colors.cyclical.IceFire,zoom=5
)
#Update layout to use an openstreetmap background and customize apperance
fig.update_layout(mapbox_style="carto-positron",title_text="Restuarant Locations",font=dict(family='Arial',size=14,color="RebeccaPurple"))

#Show the interactive map
fig.show()
```

# 2.Analyze Distribution Across Cities:

# Analyze the Geographical Distribution of Restuarants Across Different Cities

```python
#Step1 :calculate the Distribution
#Calculate the number of restaurants in each city
city_distribution=data['City'].value_counts().reset_index()
city_distribution.columns=['City','Number of Restuarants']

#Display the Distribution
print(city_distribution.head(10))
```

```python
#Step 2:Visualization of Distribution

#Bar plot for restuarant distribution across cities
plt.figure(figsize=(10,6))
sns.barplot(data=city_distribution.head(10),y='City',x='Number of Restuarants',palette="colorblind")
plt.title("Distribution of Restuarants Across Cities")
plt.xlabel("Number of Restuarants")
plt.ylabel('City')
plt.xticks(rotation=45)
plt.show()
```

# 3.Determine the Correlation Between Location and Rating:

```python
#Step1: Calculate correlation

#Calculate the qcorrelation matrix
correlation_matrix=data[['Latitude','Longitude','Aggregate rating']].corr()
#Display the correlation matrix
print(correlation_matrix)
```

```python
#Step2: Visualize Correlation

#Create a heatmap to visualize the correlation matrix
plt.figure(figsize=(8,6))
sns.heatmap(correlation_matrix,annot=True,cmap='coolwarm')
plt.title('Correlation Between Location and Rating')
plt.show()
```
